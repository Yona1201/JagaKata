## BAGIAN 2 - KUMPULAN FUNGSI (PERALATAN)

```python
import pandas as pd
import numpy as np
import torch
import re
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, multilabel_confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
from transformers import (AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments, pipeline)
from datasets import Dataset
from datetime import datetime
import json
import matplotlib.pyplot as plt
import seaborn as sns

def load_slang_dict(filepath):
    df_slang = pd.read_csv(filepath, header=None, names=['slang', 'formal'], encoding='latin-1')
    return pd.Series(df_slang.formal.values, index=df_slang.slang).to_dict()

def preprocess_text(text, slang_dictionary):
    text = str(text).lower()
    words = text.split()
    normalized_words = [slang_dictionary.get(word, word) for word in words]
    text = " ".join(normalized_words)
    text = re.sub(r'@[A-Za-z0-9_]+', ' ', text)
    text = re.sub(r'https?://[A-Za-z0-9./]+', ' ', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def augment_data(df, label_columns):
    print("\nMemulai proses augmentasi data...")
    fill_mask_pipeline = pipeline('fill-mask', model='indolem/indobertweet-base-uncased', device=0 if torch.cuda.is_available() else -1)
    minority_df = df[df[label_columns].sum(axis=1) > 0].copy()
    augmented_data = []
    for _, row in minority_df.iterrows():
        try:
            words = row['text'].split()
            if len(words) < 3: continue
            mask_index = np.random.randint(0, len(words))
            masked_words = words[:]; masked_words[mask_index] = fill_mask_pipeline.tokenizer.mask_token
            masked_text = " ".join(masked_words)
            predictions = fill_mask_pipeline(masked_text, top_k=1)
            new_text = predictions[0]['sequence']
            new_row = row.to_dict(); new_row['text'] = new_text
            augmented_data.append(new_row)
        except Exception: continue
    augmented_df = pd.DataFrame(augmented_data)
    df_balanced = pd.concat([df, augmented_df], ignore_index=True)
    print(f"Ukuran dataset setelah augmentasi: {len(df_balanced)}")
    return df_balanced

def tokenize_function(examples, tokenizer):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    sigmoid = 1 / (1 + np.exp(-logits))
    predictions = (sigmoid > 0.5).astype(int)
    f1 = f1_score(labels, predictions, average='weighted', zero_division=0)
    return {'f1': f1}

class CustomTrainer(Trainer):
    def __init__(self, *args, **kwargs):
        self.class_weights = kwargs.pop('class_weights', None)
        super().__init__(*args, **kwargs)
    def compute_loss(self, model, inputs, return_outputs=False,**kwargs):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.get("logits")
        loss_fct = torch.nn.BCEWithLogitsLoss(pos_weight=self.class_weights)
        loss = loss_fct(logits, labels.float())
        return (loss, outputs) if return_outputs else loss

print("Kumpulan Fungsi Siap.")
```

## BAGIAN 3 - FUNGSI UTAMA

```python
def jalankan_eksperimen(config, project_path):
    print("="*50); print(f"MEMULAI EKSPERIMEN REVISI (ALUR ASLI): {config['experiment_name']}"); print("="*50)

    # --- Persiapan Path ---
    REVISI_BASE_DIR = os.path.join(project_path, "revisi")
    os.makedirs(REVISI_BASE_DIR, exist_ok=True)

    DATA_PATH = os.path.join(project_path, 'data/data.csv')
    SLANG_DATA_PATH = os.path.join(project_path, 'data/data_slang_tambahan.csv')
    SLANG_DICT_PATH = os.path.join(project_path, 'data/new_kamusalay.csv')

    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    output_dir = os.path.join(REVISI_BASE_DIR, f"{timestamp}_{config['experiment_name']}")
    os.makedirs(output_dir, exist_ok=True)

    print(f"[INFO] Hasil disimpan di: {output_dir}")

    # --- 1. Load Data LAMA ---
    print("\n[1] Memuat Data...")
    df = pd.read_csv(DATA_PATH, encoding='latin-1')
    print(f"   - Data Asli: {len(df)}")

    # --- GABUNG DATA SLANG  ---
    if os.path.exists(SLANG_DATA_PATH):
        df_slang = pd.read_csv(SLANG_DATA_PATH, encoding='latin-1')
        print(f"   - Data Slang Tambahan: {len(df_slang)}")

        # Gabung jadi satu DataFrame besar
        df = pd.concat([df, df_slang], ignore_index=True)
        print(f"   - Total Data Gabungan: {len(df)}")
    else:
        print("\n[WARNING] File data_slang_tambahan.csv tidak ditemukan!")

    # --- Preprocessing ---
    slang_dict = load_slang_dict(SLANG_DICT_PATH)
    df['text'] = df['Tweet' if 'Tweet' in df.columns else 'tweet'].apply(lambda x: preprocess_text(x, slang_dict))
    label_columns = ["HS", "HS_Individual", "HS_Group", "HS_Religion", "HS_Race", "HS_Physical", "HS_Gender", "HS_Other", "HS_Weak", "HS_Moderate", "HS_Strong"]

    # --- AUGMENTASI  ---
    if config['use_augmentation']:
        print("\n[2] Melakukan Augmentasi pada SEMUA DATA (Lama + Slang)...")
        df_final = augment_data(df, label_columns)
        print(f"   - Ukuran Data Setelah Augmentasi: {len(df_final)}")
    else:
        df_final = df

    # --- SPLITTING  ---
    print("\n[3] Melakukan Splitting Data...")
    df_final['labels'] = df_final[label_columns].astype(np.float32).values.tolist()

    # Split acak dari data yang sudah diaugmentasi
    train_val_df, test_df = train_test_split(df_final, test_size=0.1, random_state=42)
    train_df, val_df = train_test_split(train_val_df, test_size=0.111, random_state=42)

    print(f"   - Final Train Size: {len(train_df)}")
    print(f"   - Final Test Size: {len(test_df)}")

    steps_per_epoch = len(train_df) // config['batch_size']
    if steps_per_epoch == 0: steps_per_epoch = 1

    # --- Hitung Bobot ---
    class_weights_tensor = None
    if config['use_cost_sensitive']:
        print("\n[4] Menghitung bobot kelas...")
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        imbalance_threshold = 5.0
        max_weight = 5.0
        use_log_smoothing = True
        custom_weights = []
        y_train_labels_df = train_df[label_columns]
        total = len(y_train_labels_df)
        for col in label_columns:
            pos = y_train_labels_df[col].sum()
            if pos == 0: w = 1.0
            else:
                neg = total - pos
                imbalance_ratio = neg / pos
                w = np.log1p(imbalance_ratio) if use_log_smoothing else imbalance_ratio
                w = 1.0 if imbalance_ratio <= imbalance_threshold else w
                w = min(w, max_weight)
            custom_weights.append(float(round(w, 4)))
        class_weights_tensor = torch.tensor(custom_weights, dtype=torch.float32).to(device)

    # --- Setup Model & Trainer ---
    print("\n[5] Menyiapkan Model & Training...")
    MODEL_NAME = "indolem/indobertweet-base-uncased"
    tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
    train_dataset = Dataset.from_pandas(train_df).map(lambda x: tokenize_function(x, tokenizer), batched=True)
    val_dataset = Dataset.from_pandas(val_df).map(lambda x: tokenize_function(x, tokenizer), batched=True)
    test_dataset = Dataset.from_pandas(test_df).map(lambda x: tokenize_function(x, tokenizer), batched=True)
    model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=len(label_columns), problem_type="multi_label_classification")

    training_args = TrainingArguments(
        output_dir=output_dir,
        num_train_epochs=config['epochs'],
        learning_rate=config['learning_rate'],
        per_device_train_batch_size=config['batch_size'],
        per_device_eval_batch_size=config['batch_size'],
        eval_steps=steps_per_epoch,
        save_steps=steps_per_epoch,
        load_best_model_at_end=False,
        metric_for_best_model="f1",
        logging_dir=f"{output_dir}/logs",
        logging_steps=10,
        seed=42,
        report_to="none"
    )
    trainer_class = CustomTrainer if config['use_cost_sensitive'] else Trainer
    trainer_kwargs = {'class_weights': class_weights_tensor} if config['use_cost_sensitive'] else {}
    trainer = trainer_class(model=model, args=training_args, train_dataset=train_dataset, eval_dataset=val_dataset,
                            tokenizer=tokenizer, compute_metrics=compute_metrics, **trainer_kwargs)

    # --- Training ---
    trainer.train()

    # --- Simpan ---
    final_model_path = os.path.join(REVISI_BASE_DIR, "MODEL_FINAL_SIDANG_REVISI")
    trainer.save_model(final_model_path)
    tokenizer.save_pretrained(final_model_path)

    # Evaluasi
    eval_results = trainer.evaluate()

    with open(os.path.join(output_dir, 'evaluation_results.json'), 'w') as f: json.dump(eval_results, f, indent=4)
    with open(os.path.join(output_dir, 'config.json'), 'w') as f: json.dump(config, f, indent=4)

    print(f"\n[SELESAI] Model Revisi tersimpan di: {final_model_path}"); print("="*50 + "\n")

print("Fungsi Utama Siap")
```
